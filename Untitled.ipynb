```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = os.getcwd() + '/tourism_csv/'
files = os.listdir('tourism_csv/')
```

```python
def standarize_columns(dataframe):
    columns = dataframe.columns.to_list()
    #convert columns to lower case remove whitspaces , joins the words with underscore '_' and remove all occurances of '.'
    columns = ["_".join(column.strip().split(" ")).lower().replace('.', '') for column in columns]
    dataframe.columns = columns
    return dataframe


def write_to_csv_files(dataframe, path, filename):
    dataframe.to_csv(path + filename,  index=False)
    return dataframe


def change_datatypes(dataframe):
    dataframe_columns = dataframe.columns
    for col in dataframe_columns.to_list()[1:]:
        try:
            dataframe[col] = pd.to_numeric(dataframe[col].str.replace(',', ''))
        except AttributeError:
            dataframe[col] = pd.to_numeric(dataframe[col])
        
    return dataframe
    
    
    
```

### Total number of tourists arrival from 1963  - 2023 per year

```python
df = pd.read_csv(path + files[9], index_col = False)
```

```python
df['total_number'] = df['total_number'] / 100000
```

```python
pre_civil_war = df[df['year'] < 1996]
civil_war = df[(df['year'] >= 1995) & (df['year'] < 2007)]
post_civil_war = df[(df['year'] >= 2006) & (df['year'] < 2020)]
covid_period = df[(df['year'] >= 2019) & (df['year'] < 2022)]
post_covid = df[df['year'] >= 2021]
```

```python
earthquake_year = df[['year', 'total_number']][df['year'] == 2015].loc[51]
```

```python
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
fig, ax = plt.subplots()
ax.plot(pre_civil_war['year'], pre_civil_war['total_number'], color='blue')
ax.plot(civil_war['year'], civil_war['total_number'], color='red', label='Civil War')
ax.plot(post_civil_war['year'], post_civil_war['total_number'], color='blue')
ax.plot(covid_period['year'], covid_period['total_number'], color='purple', label='COVID Period')
ax.plot(post_covid['year'], post_covid['total_number'], color='blue' )

ax.plot([1996, 2006], [civil_war['total_number'].iloc[0], civil_war['total_number'].iloc[-1]], color='black', linestyle='--',
        linewidth = 0.9,
        label='During Civil War')

ax.xaxis.set_minor_locator(MultipleLocator(5))
ax.grid()

ax.set_title("Number of tourist arrival in Nepal(1965 - 1923)", y = 1.05)


#plotting indian toursits

# ax.plot(total_indian['year'],total_indian['total_non_indians']/100000, '--' )
# ax.plot(total_indian['year'],total_indian['total_indian']/100000, '--' )

# Annotating the earthquake year
ax.annotate('Earthquake', xy=(earthquake_year['year'], earthquake_year['total_number']),
            xytext=(earthquake_year['year'] - 6, earthquake_year['total_number'] - 2),
            arrowprops={'facecolor': 'black', 'shrink': 0.06})

# Labeling axes and adding legend
ax.set_xlabel('Year')
ax.set_ylabel("No of tourists '000k")
ax.legend()

plt.show()
```

### Total Number of toursits by Nationality per year

```python
tourists_by_nationality = pd.read_csv('tourism_csv/11. Tourists Arrival By Major Nationalities (2014 - 2023).csv', index_col=  False)
tourists_by_nationality = tourists_by_nationality.T.reset_index(drop = False)
tourists_by_nationality.columns = tourists_by_nationality.iloc[0]
tourists_by_nationality = tourists_by_nationality.iloc[1:, :]
tourists_by_nationality = tourists_by_nationality.rename(columns = {'nationality': 'year'})
tourists_by_nationality['year'] = tourists_by_nationality['year'].astype("int32")
tourists_by_nationality = change_datatypes(tourists_by_nationality)
```

```python
import seaborn as sns
```

```python
figure = sns.heatmap(tourists_by_nationality.iloc[:, 1:], cmap='binary')
figure.set_yticklabels(labels=tourists_by_nationality['year'], rotation = 0)
figure.set_title('Tourists arrival by nationality',fontsize=12)
figure
```

<h5>It looks like after covid most of the tourists are returning but tourists from countries like japan, southkorea, myanmar and thailand aren't returning to what they were before covid</h5>

```python
no_return_country = tourists_by_nationality.columns[9:17]
nrc_df = tourists_by_nationality[no_return_country]
nrc_df.index = tourists_by_nationality['year']
```

```python
def plot_nrc(ax,data):
    
    ax.set_title("Number of tourist arrival before/after covid")
    
    ax.set_xlabel("Coutry")
    ax.set_ylabel("Year")
    for col in data.columns:
        ax.plot(data.index, data[col], label = col)
    
    ax.legend()
```

```python
fig, ax = plt.subplots()
plot_nrc(ax, nrc_df)
```

```python
fig, ax = plt.subplots(4,2, figsize = (10,8))

axes = ax.ravel()

for i, (col, ax) in enumerate(zip(nrc_df.columns, axes)):
    ax.plot(nrc_df.index, nrc_df[col],  '--',color = 'pink',)
    bar = ax.bar(nrc_df.index, nrc_df[col])
#     ax.bar_label(bar , labels = nrc_df[col])
    ax.set_title(col)
    ax.set_xlabel('Year')
    ax.set_ylabel('Value')

plt.tight_layout()
```
